````python
# GET /hello.html

import subprocess

# Command to convert Jupyter notebook to HTML
command = ["jupyter", "nbconvert", "--to", "html", "hello.ipynb"]

# Run the command
subprocess.run(command)

# Read the HTML file and print it
with open('hello.html', 'r') as file:
    print(file.read())
````
